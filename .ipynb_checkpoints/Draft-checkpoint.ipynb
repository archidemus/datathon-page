<i><h3 style="font-family:serif;font-size:200%; text-align:left;color:#4d4d4d"> I) Qué partida tiene la mayor cantidad de capitulos? se relaciona la cantidad de capitulos que tiene una partida con la cantidad de presupuesto asignado y pedido?</h3></i>

```python
df_I = pd.DataFrame(columns=('Partida','Nº de capitulos'))
i=0
for partida in df['Partida'].unique():
    df_I = df_I.append({'Partida':str(partida), 
                 'Nº de capitulos':len(df.loc[df['Partida'] == partida].Capítulo.unique())}, 
               ignore_index = True)
    i+=1
```

```python
df_I.sort_values('Nº de capitulos', ascending=False, inplace=True)

f, ax = plt.subplots(figsize = (11,9))
sn.barplot(y = df_I['Partida'].values, x = df_I['Nº de capitulos'].values, orient="h")
plt.title('Cantidad de proyectos asociados a los distintos organismos públicos', fontsize = 17)
plt.ylabel('Ministerios y organismos Públicos', fontsize = 15)
plt.xlabel('Cantidad de proyectos asociados', fontsize = 15);
```

```python
df_I_2 = pd.DataFrame(columns = ('Partida','Cantidad de proyectos','Presupuesto pedido', 'Presupuesto ejecutado'))
```

```python
for partida in df_I['Partida']:
    ppto_pedido = df.loc[df['Partida'] == partida]['Ppto_inicial_Real (millones)'].sum()
    ppto_ejecutado = df.loc[df['Partida'] == partida]['Ejec_Acum_Real (millones)'].sum()
    n_proyectos = len(df.loc[df['Partida'] == partida].Capítulo.unique())
    df_I_2 = df_I_2.append({'Partida':partida, 
                            'Cantidad de proyectos': n_proyectos,
                           'Presupuesto pedido': ppto_pedido,
                           'Presupuesto ejecutado': ppto_ejecutado},
                          ignore_index = True)
    
```

```python
# hacemos el orden por nºproyectos prioritario
df_I_2.sort_values('Cantidad de proyectos',ascending=False, inplace = True) 
df_I_2.sort_values('Presupuesto ejecutado',ascending=False)
```

```python
#df.loc[df['Ppto_inicial_Real (millones)'] == 641683894]
```

```python
f,ax = plt.subplots(figsize = (11,9))

plt.plot(df_I_2['Cantidad de proyectos'], df_I_2['Presupuesto pedido'], 'o-')
plt.plot(df_I_2['Cantidad de proyectos'], df_I_2['Presupuesto ejecutado'], 'o-')

"""
i = 0
for xy in zip(df_I_2['Cantidad de proyectos'], df_I_2['Presupuesto ejecutado']):                                       # <--
    ax.annotate('%s' % df_I_2['Partida'][i], xy=xy, textcoords='data')
    i += 1
"""
plt.title('Presupuesto pedido versus ejecutado segun la cantidad de proyectos', fontsize = 17)
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0., fontsize = 15);
```

* En general, independiente de la cantidad de proyectos que tenga el organismo público, el presupuesto pedido se condice bastante bien con el presupuesto utilizado luego de un período.
* Al parecer, los organismos con pocos capitulos tienden a ser los que más presupuesto fiscal utilizan.


<i><h3 style="font-family:serif;font-size:200%; text-align:left;color:#4d4d4d"> II) En qué años se gastó/pidió más?</h3></i>

```python
df_II = pd.DataFrame(columns = ['Periodo', 'Presupuesto pedido', 'Presupuesto ejecutado'])
for periodo in df['Periodo'].unique():
    aux_df = df.loc[df['Periodo'] == periodo]
    df_II = df_II.append({'Periodo':periodo,
                         'Presupuesto pedido': sum(aux_df['Ppto_inicial_Real (millones)']),
                         'Presupuesto ejecutado': sum(aux_df['Ejec_Acum_Real (millones)'])}, 
                         ignore_index = True)
    
    
f,ax = plt.subplots(figsize = (11,9))
plt.plot(df_II['Periodo'],df_II['Presupuesto ejecutado'], 'o-')
plt.plot(df_II['Periodo'],df_II['Presupuesto pedido'], 'o-')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0., fontsize = 15)
plt.title('Presupuesto fiscal total pedido y ejecutado a lo largo del tiempo', fontsize = 17);
```

* El presupuesto ejecutado siempre ha sido mayor al presupuesto pedido.
* El 2016 fue el año en que el *gap* entre el presupuesto total pedido y el presupuesto total ejecutado fue mayor.
* Tanto el presupuesto total pedido como el presupuesto total ejecutado por los organismos en conjunto, tienden a incrementar con el tiempo.<br>

<i><h3 style="font-family:serif;font-size:200%; text-align:left;color:#4d4d4d"> III) Qué Ministerios han gastado más a lo largo del tiempo?</h3></i>

```python
f,ax = plt.subplots(figsize = (11,9))
df_I_2.sort_values('Presupuesto ejecutado',ascending=False, inplace = True) 
sn.barplot(list(map(np.log10, df_I_2['Presupuesto ejecutado'].values)), 
           df_I_2['Partida'].values)
plt.xlabel('Presupuesto ejecutado total entre 2009 y 2017 (en log10 millones)')
plt.title('Presupuesto ejecutado de los diferentes ministerios', fontsize = 17);
```

```python
df_III = pd.DataFrame(columns = ['Partida', 'Presupuesto pedido', 'Presupuesto ejecutado', 'Periodo', 'pedido vs usado'])

for partida in df['Partida'].unique():
    for periodo in df['Periodo'].unique():
        partida_df = df.loc[(df['Partida'] == partida) & (df['Periodo'] == periodo)]
        df_III = df_III.append({'Partida': partida,
                                'Presupuesto pedido': sum(partida_df['Ppto_inicial_Real (millones)']),
                                'Presupuesto ejecutado': sum(partida_df['Ejec_Acum_Real (millones)']),
                                'Periodo': periodo}, ignore_index = True)
```

```python
######## Hacer un widget para poder cambiar el organismo publico y obtener el mismo grafico ##########

choice_part = 'TESORO PÚBLICO'
#choice_presup = 'Presupuesto ejecutado'
f,ax = plt.subplots(figsize = (11,9))
df_choice = df_III.loc[df_III['Partida'] == choice_part]
plt.plot(df_choice['Periodo'],df_choice['Presupuesto ejecutado'], 'o-')
plt.plot(df_choice['Periodo'],df_choice['Presupuesto pedido'], 'o-')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0., fontsize = 15)
plt.title('Presupuesto pedido y ejecutado a lo largo del tiempo por el %s' %choice_part, fontsize = 17);
```

```python
df_III_2 = pd.DataFrame(columns = ['Partida', 'Presupuesto pedido', 'Presupuesto ejecutado', 'pedido vs usado'])
for partida in df['Partida'].unique():
    aux_df = df.loc[df['Partida'] == partida]
    sum_pedido = sum(aux_df['Ppto_inicial_Real (millones)'])
    sum_ejecutado = sum(aux_df['Ejec_Acum_Real (millones)'])
    df_III_2 = df_III_2.append({'Partida': partida,
                                'Presupuesto pedido': sum_pedido,
                                'Presupuesto ejecutado': sum_ejecutado,
                               'pedido vs usado': sum_pedido/sum_ejecutado}, ignore_index = True)
df_III_2.sort_values(by = 'pedido vs usado', inplace = True, ascending = False)
```

```python
f,ax = plt.subplots(figsize = (11,9))
sn.barplot(y = 'Partida', x = 'pedido vs usado', data = df_III_2, orient = 'h')
plt.title('Relación entre el presupuesto pedido y el presupuesto usado por cada organismo durante el periodo 2009 a 2017', 
          fontsize = 17);
plt.axvline(x=1.0, color = 'blue');
#plt.xticks(rotation = 90)
```
